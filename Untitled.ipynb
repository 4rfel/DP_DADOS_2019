```python
import pandas as pd
```

```python
PES2015 = pd.read_csv('PES2015.csv',sep=',')
```

```python
PES2015.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
old_columns = (PES2015.columns)
```

```python
def clean_conversor_line(line, og_codes):
    new_name = ""
    first = False
    code = ""
    counter = 0
    
    for e in line:
        if e == "V":
            counter = 5
            
        if counter > 0:
            code += e
            counter -= 1
        if code in og_codes:
            if e == "/" and not first:
                first = True
            elif e != '/' and first:
                new_name += e
    
    new_name = new_name.replace("*","")
    new_name = new_name.replace("\n", "")
    
    return new_name, code


# aa = "'@00001\t\t\tV6002          $4.          /*aa*/\n'"
# clean_conversor_line(aa,old_columns)
new_columns = []
code = []
for e in conversor:
    new, c = (clean_conversor_line(e, old_columns))
    if new != "":
        new_columns.append(new)
        code.append(c)
```
